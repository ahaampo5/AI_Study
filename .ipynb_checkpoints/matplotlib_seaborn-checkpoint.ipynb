## 1. 그래프 특성

### 1.1 timeline graph
- 시간범위에 걸쳐 변화 양상을 보여줄 때 사용한다.  

### 1.2 Scatter graph
- 상관관계에 따른 분포를 보여줄 때 사용한다.  

### 1.3 Bubble graph
- 양적 분포를 강조할 때 사용한다.  

### 1.4 Bar graph
- 데이터의 크고작음을 한눈에 파악하기 쉽다. 데이터 양이 많으면 세로로, 적으면 가로로 표현하는 것이 좋다.

- 데이터 그룹에 대한 범위 및 분포 비교에 적합한 그래프다.  
- 상세 분포를 파악하기에 적합하지 않아 많이 사용하지는 않는다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<img src="./img/matplotlib.png" alt="figure" style="width: 300px;"/>

```python
# plt.show() 는 fig를 출력하는 것과 같다.
fig = plt.figure()
fig.set_size_inches(10,10)
ax = []

# 한번에 fig, ax 관리하려면 subplots 함수
fig, ax = plt.subplots(2,2)
fig.set_size_inches(10,10)
ax[0][0].plot([1,2,3],[3,4,5])
ax[1][0].plot([2,3,4],[4,5,6])

# scatter 관리
x = np.random.rand(10, 1)
area = np.pi * x * 30
ax[1][1].scatter(x,np.random.rand(10, 1),s=area,c='r', alpha=0.5)
```

```python
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot([1,2,3],[3,4,5],'#00FF11',label='l1') # linestyle(ls) /  --: 점선, ^:삼각형, o: 원
ax2.plot([2,3,4],[4,5,6], label='l2')

# xlim, ylim 대신 쓰자!
ax1.axis([0,5,0,6])

ax1.legend()
ax2.text(3, 5, 'hh')
# axes 객체로 하는 것들은 대부분 plt 함수와 공용으로 쓸 수 있다.

# ax1.title('1')
ax1.set_title('he', fontsize = 15)
# 근데 title은 좀 달라

# 기본은 연속식, 
```

## 3. Seaborn
- pandas에서 table을 가져와서 사용하기 쉽다.

### 1.1.3 Pandas를 이용하여 그리기

```python
import dateutil
df = pd.read_csv('phone_data.csv')
df.columns
```

```python
def parsedate(date_info):
    import datetime
    return date_info[:9]
```

```python
df['day'] = df['date'].apply(parsedate)
df
```

```python
plt.plot(df.index,df.duration)
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

network_types = df.network.unique().tolist()
ax = []
for i in range(1, 9):
    ax.append(fig.add_subplot(3,3,i))
    network_name = network_types[i - 1]
    
    plt.title(network_name)
    
    x_1 = df[df.network == network_name]['day']
    y_1 = df[df.network == network_name]['duration']
    ax[i - 1].get_xaxis().set_visible(False)
    ax[i - 1].plot(x_1, y_1)
```

```python
result = df.groupby(['network', 'day'])['duration'].sum().reset_index()
result.plot()
```
