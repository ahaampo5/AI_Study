## 1. 강의 정리

### 1.1 Optimization 용어 정리

#### 1.1.1 Optimizer 의 특성 용어

- Generalization  
training error는 계속 줄어들지만 train데이터의 구체적 특징을 집중적으로 학습하므로 test error는 높아지는 경향성이 생긴다. 딥러닝에서는 샘플 데이터를 가지고 전체를 포괄하는 학습을 시키는 것이므로 구체적인 것을 완벽하게 가르치는 것이 아닌 추상적인 특징을 포괄적으로 가르치는 것이 좋다고 생각할 수 있다.  
학습데이터와 테스트 데이터의 결과 차이를 의미한다. Generalization이 높아도 학습이 잘 안되어있으면 의미가 없다.

- Cross-validation (k-ford validation)  
학습데이터로 학습, validation데이터로 얼마나 학습이 되었는지 확인한다. train data 덩어리 k-1개로 학습을 시키고 나머지 하나로 잘 되어있는지 확인한다.  
최적의 하이퍼 파라메터를 찾는데도 쓴다. 최적의 파라메터를 찾은 이후에는 모든 학습데이터를 사용한다.

- Bias and Variance  
Bias : 출력의 평균이 목표에 가까운지 아는 지표  
Variance : 출력이 얼마나 좁은 분포를 가지는가를 나타내는 지표  
`Cost = bias^2 + variance + noise`  
코스트를 낮추는 것이 bias와 variance와 직접적으로 관련이 있음을 고려해야한다.

- Bootstrapping  
학습데이터 중에 일부만 활용하여 여러 모델을 만든다. 독립적으로 학습을 시키며 여러 모델을 통해 일치도를 확인할 때 쓴다.  
앙상블이라 불리기도 하며 Bagging이라 불리기도 한다.

- Boosting  
간단한 모델(Weak learner)을 여러개 만들어서 순차적으로 큰 모델(Strong learner)로 완성시켜 결과를 본다.

#### 1.1.2 Practical Optimizer

- Batch-size Matters  
배치사이즈를 작게 쓰는 것이 더 좋다. 여러 특징들의 구체적인 부분을 학습하는 것보다 추상적으로 학습시키는 것이 테스트에서는 더 좋은 결과를 가진다.    
Flat Minimum이 Sharp Minimum보다 실제 데이터에 더 효과적이다.


#### 1.1.3 Gradient Descent Method

- SGD : learning rate만을 가진다.(학습이 느리다.)  
- Momentum : learning rate에 momentum요소를 포함시켜 accumulation으로 학습시킨다.(관성을 통해 학습을 늘리는 것을 생각했다.)(하지만 관성의 성질때문에 수렴하기 어려운 특징을 가졌다.)  
![image.png](img/Momentum.png)  
- Nesterov Accelerated Gradient  
관성을 고려하지만 수렴을 하는데 지장이 없도록 lookahead gradient(관성으로 이동한 점에서 미분계산)의 개념을 도입한다.
![image.png](img/NAG.png)  
- Adagrad  
지금까지 얼마나 변해왔는지 안변해왔는지를 확인하여 조금 변해왔다면 많이 변하게 하고 많이 변해왔다면 조금 변하게 만들고 싶다.(문제점 : 시간이 지나 변한 양이 점점 커지면 학습되지 않는 현상이 생긴다.)  
![image.png](img/Adagrad.png)  
- Adadekta  
adagrad의 학습이 멈추는 현상을 개선하기 위해 만들어졌지만 하이퍼파라메터가 없기 때문에 잘 쓰이지 않는다.
![image.png](img/Adadelta.png)  
- RMSprop  
Adagrad에 EMA를 넣어 데이터가 무한정 쌓이는 것들을 방지해준다.
![image.png](img/RMSprop.png)  
- Adam  
Momentum과 Adaptive의 성질을 잘 모아 놓은 형태다.
![image.png](img/Adam.png)  

#### 1.1.4 Regularization  
개선된 Generalization이 목표

- Early stopping  
training error와 validation error를 볼 때 generalization이 늘어나는 구간에 미리 멈추는 것  
- Parameter Norm Penalty  
네트워크를 구성하는 parameter의 총크기를 줄여서 함수공간을 부드럽게 만들어준다.  
- Data Augmentation  
샘플이 전체 분포를 대표할 수 없기 때문에 가진 Dataset을 약간의 수정을 통해 데이터 셋을 늘린다.  
- Noise Robustness  
입력데이터와 weight에 noise를 넣어주어 성능을 더 높여준다.  
- Label Smoothing  
label을 섞어주어 성능을 높여준다. 예시 : Mixup(이미지섞기), Cutout(이미지를 자르기), CutMix(그림 두개를 잘라붙인다.)  
- Dropout  
몇가지의 weight를 0으로 만들어 성능을 높인다.  
- Batch Normalization  
배치를 정규분포로 분포시켜 성능을 높인다. 예시 : Batch Norm, Layer Norm, Instance Norm, Group Norm

### 1.2 Optimizer

#### 1.2.1 코드 분석

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
%matplotlib inline
%config InlineBackend.figure_format='retina'
print (f"PyTorch version:[{torch.__version__}]")
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print (f"device:[{device}].")
```

```python
n_data = 10000
x_numpy = -3+6*np.random.rand(n_data,1)
y_numpy = np.exp(-(x_numpy**2))*np.cos(10*x_numpy) + 3e-2*np.random.randn(n_data,1)
plt.figure(figsize=(8,5))
plt.plot(x_numpy,y_numpy,'r.',ms=2)
plt.show()
x_torch = torch.Tensor(x_numpy).to(device)
y_torch = torch.Tensor(y_numpy).to(device)
print ("Done.")
```

```python
class Model(nn.Module):
    def __init__(self,name='mlp',xdim=1,hdims=[16,16],ydim=1):
        super(Model, self).__init__()
        self.name = name
        self.xdim = xdim
        self.hdims = hdims
        self.ydim = ydim

        self.layers = []
        prev_hdim = self.xdim
        for hdim in self.hdims:
            self.layers.append(nn.Linear(
                # FILL IN HERE
                prev_hdim,hdim
            ))
            self.layers.append(nn.Tanh())  # activation
            prev_hdim = hdim
        # Final layer (without activation)
        self.layers.append(nn.Linear(prev_hdim,self.ydim,bias=True))

        # Concatenate all layers 
        self.net = nn.Sequential()
        for l_idx,layer in enumerate(self.layers):
            layer_name = "%s_%02d"%(type(layer).__name__.lower(),l_idx)
            self.net.add_module(layer_name,layer)

        self.init_param() # initialize parameters
    
    def init_param(self):
        for m in self.modules():
            if isinstance(m,nn.Conv2d): # init conv
                nn.init.kaiming_normal_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m,nn.Linear): # lnit dense
                nn.init.kaiming_normal_(m.weight)
                nn.init.zeros_(m.bias)
    
    def forward(self,x):
        return self.net(x)

print ("Done.")        
```

```python
LEARNING_RATE = 1e-2
# Instantiate models
model_sgd = Model(name='mlp_sgd',xdim=1,hdims=[64,64],ydim=1).to(device)
model_momentum = Model(name='mlp_momentum',xdim=1,hdims=[64,64],ydim=1).to(device)
model_adam = Model(name='mlp_adam',xdim=1,hdims=[64,64],ydim=1).to(device)
# Optimizers
loss = nn.MSELoss()
optm_sgd = optim.SGD(
    # FILL IN HERE
    lr=LEARNING_RATE,
    params=model_sgd.parameters()
)
optm_momentum = optim.SGD(
    # FILL IN HERE
    lr=LEARNING_RATE,momentum=0.9,
    params=model_momentum.parameters()
)
optm_adam = optim.Adam(
    # FILL IN HERE
    lr=LEARNING_RATE,
    params=model_adam.parameters()
)
print ("Done.")
```

```python
np.set_printoptions(precision=3)
n_param = 0
for p_idx,(param_name,param) in enumerate(model_sgd.named_parameters()):
    if param.requires_grad:
        param_numpy = param.detach().cpu().numpy() # to numpy array 
        n_param += len(param_numpy.reshape(-1))
        print ("[%d] name:[%s] shape:[%s]."%(p_idx,param_name,param_numpy.shape))
        print ("    val:%s"%(param_numpy.reshape(-1)[:5]))
print ("Total number of parameters:[%s]."%(format(n_param,',d')))
```

```python
MAX_ITER,BATCH_SIZE,PLOT_EVERY = 1e4,64,500

model_sgd.init_param()
model_momentum.init_param()
model_adam.init_param()

model_sgd.train()
model_momentum.train()
model_adam.train()

for it in range(int(MAX_ITER)):
    r_idx = np.random.permutation(n_data)[:BATCH_SIZE]
    batch_x,batch_y = x_torch[r_idx],y_torch[r_idx]
    
    # Update with Adam
    y_pred_adam = model_adam.forward(batch_x)
    loss_adam = loss(y_pred_adam,batch_y)
    optm_adam.zero_grad()
    loss_adam.backward()
    optm_adam.step()

    # Update with Momentum
    y_pred_momentum = model_momentum.forward(batch_x)
    loss_momentum = loss(y_pred_momentum,batch_y)
    optm_momentum.zero_grad()
    loss_momentum.backward()
    optm_momentum.step()

    # Update with SGD
    y_pred_sgd = model_sgd.forward(batch_x)
    loss_sgd = loss(y_pred_sgd,batch_y)
    optm_sgd.zero_grad()
    loss_sgd.backward()
    optm_sgd.step()
    

    # Plot
    if ((it%PLOT_EVERY)==0) or (it==0) or (it==(MAX_ITER-1)):
        with torch.no_grad():
            y_sgd_numpy = model_sgd.forward(x_torch).cpu().detach().numpy()
            y_momentum_numpy = model_momentum.forward(x_torch).cpu().detach().numpy()
            y_adam_numpy = model_adam.forward(x_torch).cpu().detach().numpy()
            
            plt.figure(figsize=(8,4))
            plt.plot(x_numpy,y_numpy,'r.',ms=4,label='GT')
            plt.plot(x_numpy,y_sgd_numpy,'g.',ms=2,label='SGD')
            plt.plot(x_numpy,y_momentum_numpy,'b.',ms=2,label='Momentum')
            plt.plot(x_numpy,y_adam_numpy,'k.',ms=2,label='ADAM')
            plt.title("[%d/%d]"%(it,MAX_ITER),fontsize=15)
            plt.legend(loc='upper right',fontsize=15)
            plt.show()

print ("Done.")
```

### 1.3 Convolution Neural Network

#### 1.3.1 Convolution

일반적으로 신호처리에서 Linear time-invariant system에서 input과 output이 어떤 경향을 가지고 있는지 확인하는 연산으로 사용된다. 

무선신호에서 망가지는 신호가 원신호와 얼마나 연관성이 있는지 convolution함수로 확인하여 정확한 정보를 추출하는데 쓰이는 것처럼 이미지 처리에서도 mask와 비슷한 특성을 가진 지점을 파악하는 목적으로 사용할 수 있다.  

이미지 처리에서 신기한 점은 이미지의 특성을 파악하는 것 외에 이미지의 특징을 희미하게 만드는(블러처리)까지 가능한데 이러한 특징을 보면 convolution의 목적이 두드러진 특징을 찾는 것과 동시에 추상적인 특징들을 만들어 표현 할 수 있는 능력이 있는 것 같다.
