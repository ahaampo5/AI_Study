### 1.1.2 Figure

```python
fig = plt.figure()
fig.set_size_inches(10,5)
plt.style.use('classic') # R에서 많이 쓰는 style

ax = []
colors = ['b', 'g', 'r', 'c','#79EBEB', 'y', 'k'] # hex rgb code
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i))
    x_1 = np.arange(50)
    y_1 = np.random.rand(50)
    c = colors[i]
    
    ax[i-1].plot(x_1, y_1, c=c) #plot : 그려라
plt.show()
```

```python
x_1 = np.arange(50)
y_1 = np.cos(np.arange(0,5,0.1))
x_2 = np.arange(50)
y_2 = np.sin(np.arange(0,5,0.1))
plt.title('$y = \\frac{ax + b}{test}$')
plt.plot(x_2, y_2, c='r', ls='dotted', label='line_2')
plt.plot(x_1, y_1, c="b", ls='dashed', label='line_1')
plt.grid(True, lw=0.4, ls='--', c='.50')
plt.xlim(0, 55)
plt.ylim(-1.5, 1.5)
plt.savefig('test.png')
plt.legend(loc='lower left')
```

### 1.1.3 Pandas를 이용하여 그리기

```python
import dateutil
df = pd.read_csv('phone_data.csv')
df.columns
```

```python
def parsedate(date_info):
    import datetime
    return date_info[:9]
```

```python
df['day'] = df['date'].apply(parsedate)
df
```

```python
plt.plot(df.index,df.duration)
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

network_types = df.network.unique().tolist()
ax = []
for i in range(1, 9):
    ax.append(fig.add_subplot(3,3,i))
    network_name = network_types[i - 1]
    
    plt.title(network_name)
    
    x_1 = df[df.network == network_name]['day']
    y_1 = df[df.network == network_name]['duration']
    ax[i - 1].get_xaxis().set_visible(False)
    ax[i - 1].plot(x_1, y_1)
```

```python
result = df.groupby(['network', 'day'])['duration'].sum().reset_index()
result.plot()
```
