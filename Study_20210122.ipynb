### 1.1 예외 처리

- try : 예외가 발생 할 수 있는 함수  
- except <에러 종류> : 예외가 발생하면 실행
- else : 예외가 발생하지 않으면 실행
- finally : 모든 경우에 실행
- assert : 조건이 True면 넘어가고 False면 assertion error 발생

```python
for i in range(5):
    try:
        print(5/i)
    except ZeroDivisionError as e:
        print(e)
    else:
        print('나누어졌습니다.')
    finally:
        print('--- end')
```

### 1.3 Pickle

> 객체를 파일로 저장해서 프로그램이 종료되더라도(객체의 메모리가 없어지더라도) 추후 불러올 수 있도록 바이너리로 저장.

```python
import pickle

with open('list.pickle', 'wb') as f:
    test = [1,2,3,4,5]
    pickle.dump(test, f) # test객체를 f에 저장하여라    
```

### 1.4 Logging - 로그 남기기

> 실행이나 개발 시점에서 기록을 남기는 행위

- Log 파일 생성하기

```python
import os 

if not os.path.exists('log'): # 파일 생성
    os.mkdir('log')

TARGET_FILE_PATH = os.path.join('log', 'count_log.txt')

if not os.path.exists(TARGET_FILE_PATH):
    with open('log/count_log.txt', 'w', encoding='utf8') as f:
        f.write('기록이 시작됩니다\n')
```

```python
import logging # print 와 비슷한 역할을 한다.

logging.debug('step') # 개발시 처리 기록을 남기는 경우
logging.info('information') # 서버가 시작, 종료되거나, 사용자 접속 기록 남기는 경우
logging.warning('warning') # 사용자의 잘못된 입력이나 처리는 가능하지만 의도치않은 정보가 들어오는 경우
logging.error('error') # exception 처리 후 사용자에게 알리는 경우
logging.critical('wrong!') # 강제종료 및 데이터 손실 (프로그램이 실행되지 않음)
```

```python
logger = logging.getLogger('main') # 로그 출력 레벨 조정
logging.basicConfig(level=logging.DEBUG)
logger.setLevel(logging.INFO)

logging.debug('step') # 개발시 처리 기록을 남기는 경우
logging.info('information') # 서버가 시작, 종료되거나, 사용자 접속 기록 남기는 경우
logging.warning('warning') # 사용자의 잘못된 입력이나 처리는 가능하지만 의도치않은 정보가 들어오는 경우
logging.error('error') # exception 처리 후 사용자에게 알리는 경우
logging.critical('wrong!') # 강제종료 및 데이터 손실 (프로그램이 실행되지 않음)
```
